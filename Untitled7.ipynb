```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import MinMaxScaler, StandardScaler
%matplotlib inline
```

```python
data = pd.read_csv(r'D:\Data Science\Rakamin\FinPro\dataset.csv')
```

```python
kategori = ['OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend', 'Revenue']
numerikal = ['Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration',  
             'ProductRelated', 'ProductRelated_Duration', 'BounceRates', 'ExitRates', 'PageValues','SpecialDay']
date = ['Month']
```

```python
data.duplicated().sum()
```

```python
means = data[numerikal].mean()
data[numerikal]=data[numerikal].fillna(means)
data.dropna(subset=['OperatingSystems'], inplace=True)
```

```python
# Hapus data duplikat lalu cek lagi
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
for num in data[numerikal]:
    min_value = data[data[num] > 0][num].min() / 2 
    data[num] = np.log(data[num] + min_value) 
```

```python
from scipy import stats

print("Before removing outlier: ", len(data))

for num in data[numerikal]:
    z_scores = np.abs(stats.zscore(data[num]))
    data = data[z_scores <= 3]

print("After removing outlier: ", len(data))
```
